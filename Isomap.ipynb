```python
# import Distances between cities from dataset
import pandas as pd
distances = pd.read_csv('distances.csv',index_col=0)
distances
```

```python
# convert dataframe to array
A=distances.to_numpy()
A.shape
```

```python
import numpy as np

# function for calculate matrix B from Distance matrix
def Matrix_B(A):
    
    e = np.ones((len(A), 1))
    I = np.identity(len(A))
    L = I-(1/31)*e@e.T
    
    A_tild = np.power(A, 2)
    B = (-1/2)*(L@A_tild@L)
    
    return B
```

```python
B = Matrix_B(A)
B
```

```python
def EigenValue_Decomposition(B):
    eigenvalues, Q = np.linalg.eig(B)
    i = eigenvalues.argsort()[::-1] 
    eigenvalues = eigenvalues[i]
    Q = Q[:,i]
    
    return eigenvalues, Q
```

```python
eigen_values, Q = EigenValue_Decomposition(B)
```

```python
eig_diag = np.zeros((len(eigen_values), len(eigen_values)))
for i in range(len(eigen_values)):
    eig_diag[i,i]=eigen_values[i]
    
Q@eig_diag@Q.T
```

```python
# function for calculate the position matrix from B
def MDS(A):
    B = Matrix_B(A)
    eigen_values,Q = EigenValue_Decomposition(B)
    landa1_hat = max(eigen_values[0],0)
    landa2_hat = max(eigen_values[1],0)
    l = np.array([np.sqrt(landa1_hat), np.sqrt(landa2_hat)])
    landa_k = np.diag(l)
    Q_k = Q[:,:2]
    X = Q_k@landa_k
    
    return X
```

```python
X = MDS(A)
X
```

```python
import matplotlib.pyplot as plt

# function for show cities on map
def show_map(X, k=None):
    plt.scatter(X[:, 0], X[:, 1], marker='s', c='green')
    plt.title("cities with k=%s" %k ,fontsize=13)
    for i, label in enumerate(cities):
        plt.annotate(label, (X[i, 0], X[i, 1]))
    plt.show()
```

```python
def apply_symmetry(X, symmetry_type="rotation", degree=None):
    
    if symmetry_type == "reflection":
        X_symmetry = -X
    elif symmetry_type == 'reflectionY':
        X_symmetry = X.copy()
        X_symmetry[:, 1] = -X_symmetry[:, 1]
    elif symmetry_type == 'rotation':
        rotation_angle = degree
        rotation_matrix = np.array([[np.cos(np.radians(rotation_angle)), -np.sin(np.radians(rotation_angle))],
                           [np.sin(np.radians(rotation_angle)), np.cos(np.radians(rotation_angle))]])
        X_symmetry = X.dot(rotation_matrix)
    
    return X_symmetry
```

```python
show_map(X)
X_symmetry = apply_symmetry(X , symmetry_type="reflectionY")
X_symmetry = apply_symmetry(X_symmetry , symmetry_type="rotation", degree=45)
show_map(X_symmetry)
```

```python
# neighborhood graph with k nearest neighbor
def knn_gragh(A,k):
    graph = np.zeros((len(A), len(A)))
    counter = -1
    for list in A:
        counter +=1
        ind_min = counter
        for i in range(k):
            list[ind_min] = 1e6
            min = np.amin(list)
            ind_min = np.argmin(list)
            graph[counter, ind_min] = min
            graph[ind_min, counter] = min
    return graph
```

```python
graph_k3 = knn_gragh(A,3)
graph_k3
```

```python
import networkx as nx

# function for dijkstra algorithm
def dijkstra1(graph, source):
    G = nx.Graph(graph)
    shortest_paths = nx.single_source_dijkstra_path_length(G, source)
    distances = [shortest_paths[node] for node in range(len(graph))]
    return distances
```

```python
#when the source node is 0
np.array(dijkstra1(graph_k3,0))
```

```python
# distances between any two nodes with dijkstra (Geodesic distances)
Geodesic_distance=[]
for i in range(len(graph_k3)):
    Geodesic_distance.append(np.array(dijkstra1(graph_k3,i)))
Geodesic_distance=np.array(Geodesic_distance)
Geodesic_distance
```

```python
np.shape(Geodesic_distance)
```

```python
# Isomap algorithm
# function for calculate position matrix (X)
def Isomap(A,k):
    graph_k = knn_gragh(A,k)
    Geodesic_distance=[]
    for i in range(len(A)):
        Geodesic_distance.append(np.array(dijkstra1(graph_k,i)))
    Geodesic_distance=np.array(Geodesic_distance)
    X = MDS(Geodesic_distance)
    
    return X
```

```python
# Isomap with approach knn and k=3
distances = pd.read_csv('distances.csv',index_col=0)
A=distances.to_numpy()
X_3=Isomap(A,3)
X_3.shape
```

```python
show_map(X_3, 3)
X_3_symmetry = apply_symmetry(X_3, symmetry_type="reflection")
X_3_symmetry = apply_symmetry(X_3_symmetry , symmetry_type="reflectionY")
X_3_symmetry = apply_symmetry(X_3_symmetry , symmetry_type="rotation", degree=45)
show_map(X_3_symmetry, 3)
```
